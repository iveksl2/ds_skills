```python
import datetime
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import numpy as np
```

### Simplest

```python
with PdfPages('one_plot.pdf') as pdf:
    fig, ax = plt.subplots() # Same as before. Previously this was done implicitly
    ax.plot(np.random.rand(20)
    ax.set_title('test title')
    plt.show()
    pdf.savefig(fig)
    plt.close()
```

```python
# Same as above, but explicitly 'closes' the plt & pdf. This is necessary when not using a loop or the notation above
pdf = PdfPages('simple.pdf')
plt.plot(np.random.rand(20))
plt.title('test title')
plt.show()
pdf.savefig(fig)
plt.close()
pdf.close()
```

### Multiple Plots One page

```python
pdf = PdfPages('multiple_plots_one_page.pdf')

n_rows = 2
n_cols = 2
fig, axes = plt.subplots(n_rows, n_cols)
for row_num in range(n_rows):
    for col_num in range(n_cols):
        ax = axes[row_num][col_num]
        ax.plot(np.random.rand(20))
        ax.set_title(f'Plot ({row_num+1}, {col_num+1})')
fig.suptitle('Main title')
fig.tight_layout()
plt.show()
pdf.savefig(fig)
plt.close()
pdf.close()
```

```python

"""
https://matplotlib.org/2.0.2/examples/pylab_examples/multipage_pdf.html

This is a demo of creating a pdf file with several pages,
as well as adding metadata and annotations to pdf files.
"""

# Create the PdfPages object to which we will save the pages:
# The with statement makes sure that the PdfPages object is closed properly at
# the end of the block, even if an Exception occurs.
with PdfPages('multipage_pdf.pdf') as pdf:
    plt.figure(figsize=(3, 3))
    plt.plot(range(7), [3, 1, 4, 1, 5, 9, 2], 'r-o')
    plt.title('Page One')
    pdf.savefig()  # saves the current figure into a pdf page
    plt.close()

    plt.rc('text', usetex=True)
    plt.figure(figsize=(8, 6))
    x = np.arange(0, 5, 0.1)
    plt.plot(x, np.sin(x), 'b-')
    plt.title('Page Two')
    pdf.attach_note("plot of sin(x)")  # you can add a pdf note to
                                       # attach metadata to a page
    pdf.savefig()
    plt.close()

    plt.rc('text', usetex=False)
    fig = plt.figure(figsize=(4, 5))
    plt.plot(x, x*x, 'ko')
    plt.title('Page Three')
    pdf.savefig(fig)  # or you can pass a Figure object to pdf.savefig
    plt.close()

    # We can also set the file's metadata via the PdfPages object:
    d = pdf.infodict()
    d['Title'] = 'Multipage PDF Example'
    d['Author'] = u'Jouni K. Sepp\xe4nen'
    d['Subject'] = 'How to create a multipage pdf file and set its metadata'
    d['Keywords'] = 'PdfPages multipage keywords author title subject'
    d['CreationDate'] = datetime.datetime(2009, 11, 13)
    d['ModDate'] = datetime.datetime.today()
    
```

```python
# http://blog.marmakoide.org/?p=94
pdf_pages = PdfPages('my-fancy-document.pdf')
for i in range(3):
  # Create a figure instance (ie. a new page)
  fig = plt.figure(figsize=(8.27, 11.69), dpi=100)
 
  # Plot whatever you wish to plot
  plt.plot(np.random.normal(7,7,7))  
  # Done with the page
  pdf_pages.savefig(fig)
 
# Write the PDF document to the disk
pdf_pages.close()
```
