```python
from IPython.display import YouTubeVideo
from IPython.display import HTML

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
from scipy import optimize
import typing
from typing import List
```

```python
def objective(x):
  x1 = x[0]
  x2 = x[1]
  x3 = x[2]
  x4 = x[3]
  return x1*x4*(x1+x2+x3)+x3
def constraint1(x):
  return x[0]*x[1]*x[2]*x[3]-25
def constraint2(x):
  sum_sq = 40
  for i in range(4):
    sum_sq = sum_sq - x[i]**2
  return sum_sq
```

```python
x0 = [1,5,5,1]
print(objective(x0))
```

```python
b = (1.0, 5.0)
bnds = (b,b,b,b)
con1 = {'type': 'ineq', 'fun': constraint1}
con2 = {'type': 'eq', 'fun': constraint2}
cons = [con1,con2]
```

```python
sol = minimize(objective,x0,method='SLSQP',bounds=bnds,constraints=cons)
```

```python
sol
```

```python
print('Objective Value=',sol.fun,'\nSolution values=',sol.x)
```

```python
xstart = -20
xstop = 20
increment = .1
x = np.arange(xstart,xstop,increment)
y = 2 * x*x + 20 * x - 22

plt.plot(x,y)
plt.grid

i = 0 

while y[i] > y[i+1]:
  i = i + 1

print(x[i])
print(y[i])
```

```python
def func(x):
  y = 2 * x**2 + 20*x - 22
  return y

xmin = -20
xmax = 20
dx = .1
N = int((xmax - xmin)/dx)
x = np.linspace(xmin, xmax, N+1)

y = func(x)

plt.plot(x, y)
plt.xlim([xmin,xmax])

x_min = optimize.fminbound(func, xmin, xmax)
y_min = func(x_min)

print(x_min)
print(y_min)

res = optimize.minimize_scalar(func)

print('Standard optimization library:\n', res)
```

```python
def banana(vars: List[int]):
  a = 1
  b = 100
  x, y = vars # to make code more readable
  f = (a-x)**2 + b*(y-x**2)**2
  return f

xopt = optimize.fmin(func=banana, x0=[-1.2,1])

print(xopt)
```
