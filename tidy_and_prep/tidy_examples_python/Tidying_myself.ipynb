```python
billboard = pd.read_csv('data/billboard.csv',  encoding="mac_latin2")
```

```python
billboard
```

```python
# Melting
id_vars = ["year","artist.inverted","track","time","genre","date.entered","date.peaked"]
df = pd.melt(frame=billboard,id_vars=id_vars, var_name="week", value_name="rank")
```

```python
df # This dataset is not tidy due to redundant infromation about rank & song attributes. Breaks the 'Each type of observational units. Breaks the 'Each type of observational unit forms a table' principle
 
```

```python
song_cols = ['year', 'artist.inverted', 'track', 'time', 'genre']
songs = df[song_cols].drop_duplicates()
songs = songs.reset_index(drop=True)
songs['song_id'] = songs.index
```

```python
songs
```

```python
ranks = pd.merge(df, songs, on =['year', 'artist.inverted', 'track', 'time', 'genre'])
#ranks = ranks[['song_id', 'date', 'rank']]
ranks
```
